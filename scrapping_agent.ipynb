```python
import requests
from bs4 import BeautifulSoup

def beautifulsoup_web_scrape_url(url):
    response = requests.get(url)
    soup = BeautifulSoup(response.content, 'html.parser')
    return str(soup)
```

```python
url = "https://huggingface.co/blog/train-sentence-transformers"
```

```python
data = beautifulsoup_web_scrape_url(url)
```

```python
print(data)
```

```python
pdf_url = "https://arxiv.org/pdf/2310.19923v4"
```

```python
data = beautifulsoup_web_scrape_url(pdf_url)
```

```python
print(data)
```

```python
def jinaai_readerapi_web_scrape_url(url):
  response = requests.get("https://r.jina.ai/" + url)
  return response.text
```

```python
data = jinaai_readerapi_web_scrape_url(url)
```

```python
print(data)
```

```python
from google.colab import userdata
FIRECRAWL_API_KEY = userdata.get('FIRECRAWL_API_KEY')
```

```python
import firecrawl


def firecrawl_web_scrape_url(url):
    app = firecrawl.FirecrawlApp(api_key=FIRECRAWL_API_KEY)
    scraped_data = app.scrape_url(url)["markdown"]
    return scraped_data
```

```python
data=firecrawl_web_scrape_url(url)
```

```python
print(data)
```
